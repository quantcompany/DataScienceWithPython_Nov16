## Occupations of users of a website

### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
```

### Step 3. Inspect the dataset

```python
users.head()
```

```python
users.tail()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
occupation_size = users.groupby('occupation').size()
males_by_occupation = users.groupby(['gender','occupation']).size().loc['M']
male_ratio = males_by_occupation / occupation_size
male_ratio.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').agg([min, max])['age']
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender']).mean()['age'].unstack()
```

### Step 8.  For each occupation present the percentage of women and men

```python
(100*users.groupby(['occupation','gender']).size() / occupation_size).unstack()
```
